# Helper Functions

```python
def indices_of_outliers(df, column_name, stdev_threshold):
    mask_of_outliers = np.abs(df[column_name] - df[column_name].mean()) >= (stdev_threshold*df[column_name].std())
    set_of_indices_of_outliers = df.index[mask_of_outliers].tolist()
    return set_of_indices_of_outliers

# example
indices_of_outliers(df_combined, "LotFrontage", 4)
```

```python
def remove_rows_with_outliers(df, list_of_column_names, stdev_threshold):
    outlier_index = []
    for col in list_of_column_names:
        set_of_indices_of_outliers = indices_of_outliers(df, col, stdev_threshold)
        outlier_index.extend(set_of_indices_of_outliers)
    outlier_index = list(set(outlier_index))
    df_without_outliers = df.drop(df.index[outlier_index])
    return df_without_outliers

# example
list_of_column_names = ["LotFrontage", "OverallQual", "LotArea"]
remove_rows_with_outliers(df_combined, list_of_column_names, stdev_threshold = 4)
```

# NA Imputation

```python
# check NA in original df
df_combined.isnull().sum()[df_combined.isnull().sum()>0]
```

```python
series_names = ['BsmtExposure','BsmtQual','FireplaceQu','GarageQual','GarageType','GarageYrBlt','GarageFinish','GarageCond','PoolQC','Fence','MiscFeature','Alley','BsmtCond','BsmtFinType1','BsmtFinType2']
for series_name in series_names:
    df_combined.loc[df_combined[series_name].isnull(),series_name] = 'None'
    print(df_combined[series_name].value_counts(sort = True))
    print('*'*20)
```

```python
df_combined.isnull().sum()[df_combined.isnull().sum()>0]
```

```python
df_combined['LotFrontage'] = df_combined.groupby(['Neighborhood'])['LotFrontage'].transform(lambda x: x.fillna(x.median()))
df_combined['Electrical'] = df_combined['Electrical'].fillna('SBrkr')
df_combined['MasVnrType'] = df_combined.groupby(['Neighborhood'])['MasVnrType'].transform(lambda x: x.fillna(x.mode()[0]))
# Where there is no MasVnrType, assign the missing values of area to be zero
mask_1 = (df_combined["MasVnrType"] == "None") & (df_combined["MasVnrArea"].isna())
df_combined.loc[mask_1, "MasVnrArea"]  = 0.0
# There is 1 house with a "stone" type MasVnrType, with no area, so we impute using median area for stone types
df_combined['MasVnrArea'] = df_combined.groupby(['MasVnrType'])['MasVnrArea'].transform(lambda x: x.fillna(x.median()))
# There is one house in test set that containing missing value in GarageArea, seems like it does not have a Garage
df_combined['GarageArea']= df_combined['GarageArea'].fillna(0)
df_combined['GarageCars']= df_combined['GarageCars'].fillna(0)
# Only one house has missing value in BsmtSF due to no basement, should fill NA with 0
df_combined['TotalBsmtSF'] = df_combined['TotalBsmtSF'].fillna(0)
df_combined['BsmtFinSF1'] = df_combined['BsmtFinSF1'].fillna(0)
df_combined['BsmtFinSF2'] = df_combined['BsmtFinSF2'].fillna(0)
df_combined['BsmtUnfSF'] = df_combined['BsmtUnfSF'].fillna(0)
# There are 2 houses that have missing values in basement bathroom due to no basement 
df_combined['BsmtFullBath'] = df_combined['BsmtFullBath'].fillna(0)
df_combined['BsmtHalfBath'] = df_combined['BsmtHalfBath'].fillna(0)
df_combined['Exterior1st'] = df_combined.groupby(['RoofMatl'])['Exterior1st'].transform(lambda x: x.fillna(x.mode()[0])) 
df_combined['Exterior2nd'] = df_combined.groupby(['RoofMatl'])['Exterior2nd'].transform(lambda x: x.fillna(x.mode()[0])) 
df_combined['KitchenQual'] = df_combined.groupby(['OverallCond'])['KitchenQual'].transform(lambda x: x.fillna(x.mode()[0])) # group by Overall Condition
# 2 houses have missing values in Functional. Both are in poor quality condition and sold as abnormal type. Should not consider it as Functional_typ
df_combined['Functional'] = df_combined['Functional'].fillna('Mod')
# 4 houses having missing values in MSZoning
df_combined['MSZoning'] = df_combined.groupby(['Neighborhood'])['MSZoning'].transform(lambda x:x.fillna(x.mode()[0]))
df_combined['Utilities'] = df_combined['Utilities'].fillna('AllPub')
df_combined['SaleType'] = df_combined.groupby(['SaleCondition'])['SaleType'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df_combined.isnull().sum()[df_combined.isnull().sum()>0]
```

# Create Custom Features

```python
# Create Custom Feature: Adjusted Total Basement Area
df_combined['AdjTotalBsmtSF'] = df_combined['BsmtFinSF1']+df_combined['BsmtFinSF2']+0.9*df_combined['BsmtUnfSF']
# Create Custom Feature: Adjusted Total Bathroom 
df_combined['AdjTotalBath'] = df_combined['FullBath']+df_combined['BsmtFullBath']+0.6*(df_combined['HalfBath']+df_combined['BsmtHalfBath'])
# Create Custom Feature: HouseAge
df_combined['HouseAge'] = df_combined['YrSold']-(0.6*df_combined['YearRemodAdd']+0.4*df_combined['YearBuilt'])
# Custom Feature: AdjOutdoorSF
df_combined['AdjOutdoorSF'] = df_combined['3SsnPorch']+df_combined['EnclosedPorch']+df_combined['OpenPorchSF']+df_combined['ScreenPorch']+df_combined['WoodDeckSF']

df_combined.shape
df_combined.to_csv("combined.csv",index=False)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_categorical_le = df_categorical.apply(le.fit_transform)
df_categorical_le.head(5)
```

### Combining Label Encoded Categorical Columns and Numeric Columns

```python
df_numeric = df_combined._get_numeric_data() # need to impute missing value for MasVnrArea
numeric_columns = df_numeric.columns.to_list()
```

```python
df_le = pd.DataFrame(np.concatenate([df_numeric,df_categorical_le],axis = 1))
df_le.columns = numeric_columns + categorical_columns
df_le
```

```python
X = df_le.iloc[:1460,:].drop(['Id','SalePrice'],axis =1)
Y = df_le.iloc[:1460,]['SalePrice']
X1 = df_le.iloc[1460:,].drop(['Id','SalePrice'],axis =1) # Original Train dataset
```

### Gradient Boosting Model

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

gbm = GradientBoostingRegressor()
gbm.fit(X,Y)
gbm.score(X,Y)
```

```python
gbm.set_params(verbose=1)
gbm.fit(X,Y) 
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
# a train test split of 20% test set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,random_state = 42)
```

#### Bagging

```python
gbm.set_params(learning_rate=0.03, n_estimators=800, verbose=1, subsample=0.7, warm_start=False, loss='ls', max_depth=2, max_features=2, min_impurity_decrease=0.01)
# subsample = 0.7: each tree only use random 70% of the whole dataset for bagging
```

```python
gbm.fit(X_train,Y_train)
print('-'*50)
print("The train set R^2 is %.3f" %(gbm.score(X_train, Y_train)))
print("The test set R^2 is %.3f" %(gbm.score(X_test,Y_test)))
print("The difference between test and train is %.3f" %(gbm.score(X_train, Y_train)-gbm.score(X_test,Y_test)))
```

### Rank the Feature Importance and Select only the top 35 to refit gbm model 

```python
sorted_importance = sorted(zip(X.columns.to_numpy(), gbm.feature_importances_), key=lambda t:t[1], reverse=True)
sorted_importance_top35 = sorted_importance[:35]
sorted_importance_top35
```

```python
a, b = zip(*sorted_importance_top35) # this splits the list of tuples into two lists
pd.DataFrame({'feature_name':a, 'importance_score':b}).plot.bar(x='feature_name',y='importance_score', legend=None)
plt.title('Feature Importance Plot of 500-Tree GBM')
plt.rcParams['figure.figsize'] = (10,15)
```

```python
X_top35 = X.loc[:,list(a)]
X_top35.head(5)
```

```python
from sklearn.model_selection import train_test_split
# a train test split of 20% test set
X_top35_train, X_top35_test, Y_top35_train, Y_top35_test = train_test_split(X_top35,Y, test_size=0.2,random_state = 42)
```

```python
gbm_top35 = GradientBoostingRegressor()
gbm_top35.set_params(learning_rate=0.03, n_estimators=600, verbose=1, subsample=0.7, warm_start=False, loss='ls', max_depth=2, max_features=2, min_impurity_decrease=0.01)
# subsample = 0.7: each tree only use random 70% of the whole dataset for bagging
```

```python
gbm_top35.fit(X_top35_train,Y_top35_train)
print('-'*50)
print("The train set R^2 is %.3f" %(gbm_top35.score(X_top35_train, Y_top35_train)))
print("The test set R^2 is %.3f" %(gbm_top35.score(X_top35_test,Y_top35_test)))
a = gbm_top35.score(X_top35_train, Y_top35_train)
b = gbm_top35.score(X_top35_test,Y_top35_test)
print("Difference between train and test set R^2 is %.3f" % (a-b))
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
grid_para_tree = [{
    'n_estimators':np.arange(100,1000,100),
    'learning_rate':np.linspace(0.02,0.05,5),
#     "min_samples_leaf": range(1, 10, 3),
#     'subsample': np.linspace(0.6,0.9,4),
    "min_samples_split": np.linspace(start=2, stop=30, num=7, dtype=int) # not recommend to do a through research, try num =7 first 
}]

gbm_top35.set_params(random_state=108)
grid_search_tree = GridSearchCV(gbm_top35, grid_para_tree, cv=3, n_jobs=6)
%time grid_search_tree.fit(X_top35_train, Y_top35_train)
```
